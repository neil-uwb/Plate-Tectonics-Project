## Imports

```python
import dateutil.parser
import pandas as pd
import matplotlib.pyplot as plt
import requests
import matplotlib.colors

from PIL import Image
import urllib.request
import io
```

## Imported Data Sets from Github

```python
data = pd.read_csv('https://raw.githubusercontent.com/neil-uwb/Plate-Tectonics-Project/master/query.csv')
data2 = pd.read_csv('https://raw.githubusercontent.com/neil-uwb/Plate-Tectonics-Project/master/query2.csv')
```

## Imported Picture for the Graph

```python
url = 'https://raw.githubusercontent.com/neil-uwb/Plate-Tectonics-Project/master/Map.PNG'

img = plt.imread(url)
```

## Data Array for 2010-2020

```python
time = []
lat = []
long = []
mag = []

for i in range(len(data)):
    time.append(dateutil.parser.isoparse(data['time'][i]))
    lat.append(data['latitude'][i])
    long.append(data['longitude'][i])
    mag.append(data['mag'][i])
```

## Data Array for April 2015

```python
time2 = []
lat2 = []
long2 = []
mag2 = []

for i in range(len(data2)):
    time2.append(dateutil.parser.isoparse(data2['time'][i]))
    lat2.append(data2['latitude'][i])
    long2.append(data2['longitude'][i])
    mag2.append(data2['mag'][i])
```

## Bounding Box for 2010-2020

```python
BBox = ((data.longitude.min(),data.longitude.max(),data.latitude.min(),data.latitude.max()))
```

## Plot for 2010-2020

```python
df = pd.DataFrame({'x':long,'y':lat,'mag':mag})
cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=2.5, vmax=6)

plt.scatter(df.x, df.y, color=cmap(norm(df.mag.values)),alpha=.2,s=10)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
plt.colorbar(sm, label = 'Magnitude')
plt.xlim(BBox[0],BBox[1])
plt.ylim(BBox[2],BBox[3])
plt.xticks(rotation=30)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earthquake Data from 2010-2020 around Juan de Fuca Plate')
plt.imshow(img, zorder=0, extent = BBox, aspect= 'equal')
plt.gcf().set_size_inches(20, 10)
plt.show()
```

## Bounding Box for April 2015

```python
BBox2 = ((data2.longitude.min(),data2.longitude.max(),data2.latitude.min(),data2.latitude.max()))
```

## Plot for April 2015

```python
df = pd.DataFrame({'x':long2,'y':lat2,'mag':mag2})
cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=2.5, vmax=6)

plt.scatter(df.x, df.y, color=cmap(norm(df.mag.values)),alpha=1,s=10)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
plt.colorbar(sm, label = 'Magnitude')
plt.xlim(BBox2[0],BBox2[1])
plt.ylim(BBox2[2],BBox2[3])
plt.xticks(rotation=30)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earthquake Data from April 2015 around Juan de Fuca Plate')
plt.imshow(img, zorder=0, extent = BBox, aspect= 'equal')
plt.gcf().set_size_inches(20, 10)
plt.show()
```

```python
df = pd.DataFrame({'x':long,'y':lat,'mag':mag})
cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=2.5, vmax=6)

plt.scatter(df.x, df.y, color=cmap(norm(df.mag.values)),alpha=.2,s=10)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
plt.colorbar(sm, label = 'Magnitude')
plt.xlim(BBox[0],BBox[1])
plt.ylim(BBox[2],BBox[3])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(-127.8140,-124.10)
plt.ylim(39.97,40.70)
plt.xticks(rotation=30)
plt.title('Earthquake Data from 2010-2020 around Juan de Fuca Plate (Divergent only)')
plt.imshow(img, zorder=0, extent = BBox, aspect= 'equal')
plt.gcf().set_size_inches(10, 2)
plt.show()
```
